### Step 4: 
__As per the new govt. policy, all citizens above age 60 should not be made to work more than 25 hours per week. Let us look at the data and see if that policy is followed.__

```python
senior_citizen=census[census[0:,0]>60]
senior_citizen
```

```python
working_hours_sum=senior_citizen.sum(axis=0)[6]
print("Total working hours of senior citizen are ",working_hours_sum)
```

```python
senior_citizen_len=len(senior_citizen)
print("number of senior citizen is",senior_citizen_len)
```

```python
avg_working_hours=working_hours_sum/senior_citizen_len
print("average working hours of senior citizen",avg_working_hours)
```

__Average working hours for senior citizen is more than 25 hours per week .
Govt. policy is not followed__


### Step 5: 
__Our parents have repeatedly told us that we need to study well in order to get a good(read: higher-paying) job. Let's see whether the higher educated people have better pay in general.__

```python
high=census[census[:,1]>10]
high
```

```python
low=census[census[:,1]<=10]
low
```

```python
avg_pay_high=high[:,7].sum()/len(high)
print("average high pay",avg_pay_high)
```

```python
avg_pay_low=low[:,7].sum()/len(low)
print("average low pay:",avg_pay_low)
```

#### higher educated peoples have better pay in general
